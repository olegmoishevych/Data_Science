```python
import numpy as np

x = np.array([[1, 50], [1, 60], [1, 70], [1, 100]])
```

```python
x
```

```python
y = np.array([[300000], [350000], [400000], [500000]])
```

```python
y
```

```python
# Транспонирование матрицы x
x_t = x.T
```

```python
# Умножение транспонированной матрицы x на саму матрицу x
result = np.dot(x_t, x)
```

```python
xtx = np.dot(x_t, x)

# Нахождение обратной матрицы (X^T X)^-1
xtx_inv = np.linalg.inv(xtx)

# Умножение транспонированной матрицы X на вектор y
xty = np.dot(x_t, y)

# Вычисление коэффициентов w
w = np.dot(xtx_inv, xty)
```

```python
from matplotlib import pyplot as plt

# Построение графика
plt.scatter(x[:, 1], y, color='blue', label='Данные')

# Линия регрессии
x_line = np.linspace(40, 110, 100)  # Площади домов для предсказаний
y_line = w[0] + w[1] * x_line       # Предсказанные цены домов

plt.plot(x_line, y_line, color='red', label='Линия регрессии')

# Настройка графика
plt.xlabel('Площадь дома (кв. м)')
plt.ylabel('Цена дома (доллары)')
plt.title('Линейная регрессия: цена дома vs площадь')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
temperatures = np.array([20, 25, 30, 35, 40])
sales = np.array([200, 250, 300, 350, 400])

# Создание матрицы признаков X (сдвиговый член и температуры)
X = np.array([[1, temp] for temp in temperatures])

# Создание вектора целевых значений y (продажи мороженого)
y = sales.reshape(-1, 1)

# Транспонирование матрицы X
X_t = X.T

# Умножение транспонированной матрицы X на саму матрицу X
XtX = np.dot(X_t, X)

# Нахождение обратной матрицы (X^T X)^-1
XtX_inv = np.linalg.inv(XtX)

# Умножение транспонированной матрицы X на вектор y
Xty = np.dot(X_t, y)

# Вычисление коэффициентов w
w = np.dot(XtX_inv, Xty)

# Вывод коэффициентов
print("Коэффициенты w (или β):")
print(w)

# Функция для предсказания продаж мороженого по температуре
def predict_sales(temp, coefficients):
  return coefficients[0] + coefficients[1] * temp

# Пример предсказания для температуры 32 градуса Цельсия
predicted_sales = predict_sales(32, w)
print(f"Предсказанные продажи мороженого при температуре 32°C: {predicted_sales[0]:.2f} тысяч штук")

# Построение графика
plt.scatter(temperatures, sales, color='blue', label='Данные')

# Линия регрессии
temp_line = np.linspace(15, 45, 100)  # Температуры для предсказаний
sales_line = w[0] + w[1] * temp_line  # Предсказанные продажи

plt.plot(temp_line, sales_line, color='red', label='Линия регрессии')

# Настройка графика
plt.xlabel('Температура (°C)')
plt.ylabel('Продажи мороженого (тыс. штук)')
plt.title('Линейная регрессия: продажи мороженого vs температура')
plt.legend()
plt.grid(True)
plt.show()
```
