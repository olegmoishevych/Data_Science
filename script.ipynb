```python
import numpy as np
import matplotlib.pyplot as plt

# Пример данных
x_hw = np.array([1, 2, 3, 4, 5])
y_hw = np.array([2, 3, 5, 7, 11])

# Визуализация набора точек
plt.scatter(x_hw, y_hw, color='blue', label='Data points')
plt.xlabel('x_hw')
plt.ylabel('y_hw')
plt.title('Scatter plot of x_hw vs y_hw')
plt.legend()
plt.show()

# Построение коэффициентов регрессии
n = len(x_hw)
x_mean = np.mean(x_hw)
y_mean = np.mean(y_hw)

u1 = np.sum((x_hw - x_mean) * (y_hw - y_mean)) / np.sum((x_hw - x_mean) ** 2)
u0 = y_mean - u1 * x_mean

print(f'Коэффициенты регрессии: u0 = {u0}, u1 = {u1}')

# Предсказание
y_pred = u0 + u1 * x_hw

# Визуализация предсказания
plt.scatter(x_hw, y_hw, color='blue', label='Data points')
plt.plot(x_hw, y_pred, color='red', label='Regression line')
plt.xlabel('x_hw')
plt.ylabel('y_hw')
plt.title('Regression line with data points')
plt.legend()
plt.show()
```
