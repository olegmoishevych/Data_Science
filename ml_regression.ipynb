```python
import matplotlib.pyplot as plt

# Данные
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Создание диаграммы разброса
plt.scatter(x, y)

# Настройка меток осей
plt.xlabel('X - Годы')
plt.ylabel('Y - Доход')

# Заголовок графика
plt.title('Доход компании по годам')

# Отображение графика
plt.show()
```

```python
import numpy as np

# Создание квадратной матрицы
A = np.array([[1, 2], [3, 4]])

# Вычисление обратной матрицы
A_inv = np.linalg.inv(A)

# Вывод результатов
print("Матрица A:")
print(A)

print("Обратная матрица A_inv:")
print(A_inv)

# Проверка результата
I = np.dot(A, A_inv)
print("Произведение A и A_inv (единичная матрица):")
print(I)
```
